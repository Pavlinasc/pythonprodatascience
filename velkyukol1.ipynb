```python
#Bonus 1
import pandas as pd
import statsmodels.formula.api as smf
data = pd.read_csv("df.csv")
formula = "Y ~ CR + DA + FL + OPM + RT + TAT + DE + PT"
mod = smf.rlm(formula=formula, data=data)
res = mod.fit()
res.summary()

#sledujeme p>|z| aby bola pod 0.05. Tym padom v tomto modele mame CR, FL, OPM a TAT statisticky vyznamne
```

```python
#Bonus 2

from statsmodels.stats.outliers_influence import OLSInfluence
formula = "Y ~ CR + DA + FL + OPM + RT + TAT + DE + PT"
mod = smf.ols(formula=formula, data=data)
results = mod.fit()
influence = OLSInfluence(results)
data['Cook Distance'] = influence.cooks_distance[0]
data.head()

data = data[data['Cook Distance'] < 1]
mod = smf.ols(formula=formula, data=data)
res = mod.fit()
res.summary()
```

```python
data["Odhad_Y"] = res.predict(data)
data["Rozdil_Y"] = data["Y"] - data["Odhad_Y"]

max = data.loc[data["Rozdil_Y"].idxmax()]
max
#pouzivam z predchadzajucej ulohy, neviem ci je naozaj spravne :)
```
